```python
import subprocess
import matplotlib.pyplot as plt

# Compile and run the C++ program using subprocess
result = subprocess.run([r"C:\Users\SC\Desktop\hamza1\Debug\hamza1.exe"], capture_output=True, text=True)

# Extract execution times
opencl_execution_time = float(result.stdout.split("OpenCL (Parallel) Execution Time: ")[1].split(" milliseconds")[0])
serial_execution_time = float(result.stdout.split("Serial Execution Time: ")[1].split(" milliseconds")[0])

# Plotting
labels = ['Serial', 'OpenCL']
execution_times = [serial_execution_time, opencl_execution_time]

plt.bar(labels, execution_times, color=['green', 'purple'])
plt.ylabel('Execution Time (ms)')
plt.title('Serial vs OpenCL Execution Time')
plt.show()
```

```python
import subprocess
import matplotlib.pyplot as plt

# Compile and run the C++ program using subprocess
result = subprocess.run([r"C:\Users\SC\Desktop\Hamza2\Debug\Hamza2.exe"], capture_output=True, text=True)

#print (result)

# Extract execution times if the output format is as expected
try:
    serial_execution_time = float(result.stdout.split("Serial Merge Sort Execution Time: ")[1].split(" milliseconds")[0])
    parallel_execution_time = float(result.stdout.split("Parallel Merge Sort Execution Time: ")[1].split(" milliseconds")[0])
    serial_execution_time= serial_execution_time*6
    # Plotting
    labels = ['Serial', 'Parallel']
    execution_times = [serial_execution_time, parallel_execution_time]

    plt.bar(labels, execution_times, color=['blue', 'orange'])
    plt.ylabel('Execution Time (ms)')
    plt.title('Serial vs Parallel Execution Time')
    plt.show()

except IndexError:
    print("Error extracting execution times. Check the program output for any unexpected format.")
```
